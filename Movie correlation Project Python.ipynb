```python
# import libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize']=(12,8) # Adjusts the configuration of the plote we create


# read in the data

df = pd.read_csv(r'C:\Users\shiva\Downloads\movies.csv')
```

```python
#lookig at the data 

df.head()
```

```python
# any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
df=df.dropna()
```

```python
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# data types of our columns

df.dtypes
```

```python
#change data types of columns

df['budget']=df['budget'].astype('int64')

df['gross']=df['gross'].astype('int64')

df['votes']=df['votes'].astype('int64')

df['runtime']=df['runtime'].astype('int64')

df['score']=df['score'].astype('int64')
```

```python
# extracting the year from 'released'

df['yearcorrect']=df['released'].str.extract(pat='([0-9]{4})').astype(str)
```

```python
# Order our Data a little bit to see

df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# scatter plot with budget vs gross


plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Gross Earnings')

plt.ylabel('Budget for film')

plt.show()
```
